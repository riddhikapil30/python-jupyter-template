### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)  # [1, 2, 3, 4, 5]
y = x ** 2           # [1, 4, 9, 16, 25]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_y = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling: Ensure data arrays are non-empty and properly matched
if (
    len(x) > 0 and len(y) > 0 and len(x) == len(y) and
    len(categories) > 0 and len(values) > 0 and len(categories) == len(values) and
    len(labels) > 0 and len(sizes) > 0 and len(labels) == len(sizes)
):
    # Create 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # First subplot: Line Plot
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label="y = x²")
    axes[0, 0].set_title("Line Plot")
    axes[0, 0].set_xlabel("X-axis")
    axes[0, 0].set_ylabel("Y-axis")
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Second subplot: Bar Plot
    axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
    axes[0, 1].set_title("Bar Plot")
    axes[0, 1].set_xlabel("Categories")
    axes[0, 1].set_ylabel("Values")

    # Third subplot: Scatter Plot
    axes[1, 0].scatter(x, scatter_y, color='r', marker='o', label="Data Points")
    axes[1, 0].set_title("Scatter Plot")
    axes[1, 0].set_xlabel("X-axis")
    axes[1, 0].set_ylabel("Y-axis")
    axes[1, 0].legend()

    # Fourth subplot: Pie Chart
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True)
    axes[1, 1].set_title("Pie Chart")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show all plots
    plt.show()
else:
    print("Error: One or more data arrays are empty or mismatched in length.")
```
